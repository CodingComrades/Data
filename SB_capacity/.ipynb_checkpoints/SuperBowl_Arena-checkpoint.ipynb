```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

```python
capacity_data = "../SB_capacity/Superbowl_Arena_Capacity.csv"
capacity_df = pd.read_csv(capacity_data)
capacity_df.head(100)
```

```python
plt.figure(figsize=(50,100))
x_values = capacity_df["Year"]

plt.subplot(311)
plt.title("Superbowl Attendees 1967-2018", fontsize=20)
plt.xticks(rotation=45)
plt.xlabel("Year")
plt.ylabel("Number of Attendees")
plt.plot(x_values, capacity_df["Attendees"], label="Attendees")
plt.legend()
plt.xlim(0, len(x_values))
plt.grid(axis="y")

plt.subplot(312)
plt.title("Superbowl Capacity 1967-2018", fontsize=20)
plt.xticks(rotation=45)
plt.xlabel("Year")
plt.ylabel("Number of Attendees")
plt.plot(x_values, capacity_df["Capacity"], label="Capacity", color="Red")
plt.legend()
plt.xlim(0, len(x_values))
plt.grid(axis="y")

plt.subplot(313)
plt.title("Superbowl Attendees vs Capacity 1967-2018", fontsize=20)
plt.xticks(rotation=45)
plt.xlabel("Year")
plt.ylabel("Number of Attendees")
plt.plot(x_values, capacity_df["Capacity"], label="Capacity", color="Red")
plt.plot(x_values, capacity_df["Attendees"], label="Attendees")
plt.legend()
plt.xlim(0, len(x_values))
plt.grid(axis="y")
```
