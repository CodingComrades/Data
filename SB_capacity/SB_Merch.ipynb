```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

```python
merch_data = "../SB_capacity/Superbowl_Merchandise.csv"
merch_df = pd.read_csv(merch_data)
merch_df.head(21)
```

```python
plt.figure(figsize=(50,100))
x_values = merch_df["Year"]

plt.title("Superbowl Shirts 2000-2020", fontsize=20)
plt.xticks(rotation=45)
plt.xlabel("Year")
plt.ylabel("Price of Shirts")
plt.plot(x_values, merch_df["Shirts"], label="Shirts")
plt.plot(x_values, merch_df["Hats"], label="Hats", color="Green")
plt.plot(x_values, merch_df["Hoodies"], label="Hoodies", color="Red")
plt.legend()
plt.grid(axis="y")

```
